```python
%%time
label_counts = collections.defaultdict(list)
label_version_counts = collections.defaultdict(list)
label_date_counts = collections.defaultdict(list)
label_version_date_counts = collections.defaultdict(list)
for article in mongo['article'].find({'tags': {'$ne': []}}, {'tags': 1, 'date': 1}):
    for tag in article['tags']:
        if isinstance(tag['label'], list):
            continue
        label_counts[tag['label']].append(article['_id'])
        if 'version' in tag:
            label_version_counts[(tag['label'], tag['version'])].append(article['_id'])
        if 'date' in tag:
            label_date_counts[(tag['label'], tag['date'].year)].append(article['_id'])
        if 'version' in tag and 'date' in tag:
            label_version_date_counts[(tag['label'], tag['version'], tag['date'].year)].append(article['_id'])
```

```python
def summarize():
    
    summary = collections.defaultdict(list)
    
    for article in mongo['article'].find({'tags': {'$ne': []}}, {'tags': 1, 'date': 1}):
        
        for tag in article['tags']:
            
            if isinstance(tag['label'], list):
                continue

            _id = article['_id']
            label = tag['label']
            version = tag.get('version')
            date = article['date'].year if article['date'] else None

            summary[(label,)].append(_id)
            summary[(label, date)].append(_id)
            if version:
                summary[(label, version)].append(_id)
                summary[(label, date, version)].append(_id)
    
    return summary

%time summary = summarize()
```

```python
%%time
import collections
pipelines = collections.defaultdict(int)
pipelines_year = collections.defaultdict(lambda: collections.defaultdict(int))
total_year = collections.defaultdict(int)

pipelines_group = collections.defaultdict(
    lambda: collections.defaultdict(int)
)

pipelines_group_year = collections.defaultdict(
    lambda: collections.defaultdict(
        lambda: collections.defaultdict(int)
    )
)

variance_groups = {
    name: group.labels 
    for name, group in pattern.tag_groups.items()
}

def cat_labels(names, groups):
    return sum([
        groups[name].labels
        for name in names
    ], [])

variance_groups['all'] = cat_labels(
    pattern.tag_groups.keys(),
    pattern.tag_groups
)

variance_groups['collection'] = cat_labels(
    ['mag', 'pulse', 'task', 'des'],
    pattern.tag_groups
)

variance_groups['analysis'] = cat_labels(
    ['pkg', 'proc', 'mod', 'mcc', 'tech'], 
    pattern.tag_groups
)

for article in mongo['article'].find({'tags': {'$ne': []}}, {'tags': 1, 'date': 1}):
    
    date = article.get('date')
    year = date.year if date else None
    
    #group_data = collections.defaultdict(list)
    group_data = collections.defaultdict(tuple)
    
    total_year[year] += 1
    
    for tag in article['tags']:
        
        frozen_tag = [
            (key, value)
            for key, value in tag.items()
            if key != 'context'
                and not isinstance(value, list)
        ]
        
        for group_name, labels in variance_groups.items():
            if tag['label'] in labels:
                #group_data[group_name].append(frozen_tag)
                group_data[group_name] += tuple(frozen_tag)
    
    for group_name in variance_groups.keys():
        #frozen_tags = frozenset(group_data[group_name])
        frozen_tags = group_data[group_name]
        if frozen_tags:
            pipelines_group[group_name][frozen_tags] += 1
            pipelines_group_year[group_name][year][frozen_tags] += 1
```

```python

def summarize_custom(articles, func):

    summary = collections.defaultdict(list)

    for article in articles:

        _id = article['_id']

        for tag in article['tags']:

            out = func(tag)
            date = article['date'].year if article['date'] else None

            if out is not None:
                summary[(out,)].append(_id)
                summary[(out, date)].append(_id)
    
    return summary

def field_summary(tag):

    if tag['label'] == 'field_strength':
        return round(tag['value'] * 2.0) / 2
```

```python
articles = mongo['article'].find({'tags': {'$ne': []}}, {'tags': 1, 'date': 1})
%time s = summarize_custom(articles, field_summary)
```

```python
steps = ['realign', 'norm', 'spatsmoo', 'strip', 'stc']
articles = mongo['article'].find({'tags.label': {'$in': steps}})
sequences = []
sequence_counts = collections.defaultdict(int)
for article in articles:
    sequence = []
    for tag in article['tags']:
        if tag['label'] in steps and 'span' in tag:
            sequence.append((tag['label'], sum(tag['span']) / 2.0))
    sequence = [
        step[0]
        for step in sorted(sequence, key=lambda item: key[1])
    ]
    sequences.append(sequence)
    sequence_counts[tuple(sequence)] += 1
```
